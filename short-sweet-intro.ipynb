```python
# we've discovered a shortcoming of lists... They're a list of single pieces of info...
# Dictionary is like a labeled list
# Real World example is Nutrition Facts chart on cereal
# Curly braces start a dictionary

cereal = {
    "name": "Chex",
    "company": "General Mills",
    "sugar": 16,
    "carbs": 32,
    "protein": 2
}
```

```python
cereal
```

```python
# "key": "value" pairs separated by commas
fruit = {
    "name": "apple",
    "kind": "cosmic",
    "sugar": 8,
    "protein": 0,
    "carbs": 10
}
```

```python
fruit
```

```python
breakfast = [cereal, fruit]
breakfast
```

```python
# A spreadsheet or database table is a list of lists
# A spreadsheet or database table could also be a list of dictionaries (b/c the dictionaries have the columns names inside them already)
# Loop across the list of dictionaries to produce some totals
total_sugar = 0
total_carbs = 0
total_protein = 0

# The "item" variable didn't exist before
# "item" will hold each and every dictionary
for item in breakfast:
  # += works like a running total or a tab
  # += is an incrementing operator 
  total_sugar += item["sugar"]
  total_carbs += item["carbs"]
  total_protein += item["protein"]
```

```python
# The variables have a name, but we dont' have a label... hmm...
total_sugar, total_carbs, total_protein
```

```python
# The dictionary labels are PART of the data now, not just the variable names
breakfast_nutrition = {
    "sugar": total_sugar,
    "carbs": total_carbs,
    "protein": total_protein
}
breakfast_nutrition
```

## Where does "Data Cleaning" come in? Why is it important?

```python
# Data cleaning is critical b/c real world data is messy, ugly, incomplete, but it's valuable
# If you delete the hashtag at the beggining of the following line, you'll see an error from dealing with "messy" data
# min([1, 2, "3", "3.25", "$4.99", "$2,222"])
```

```python
# Float function converts strings into numbers (floating point numbers)
float("1.5") + 2
```

```python
# Example Data Cleaning Process at the top of the hour :)
tea = "$1.50"
eggrolls = 4
noodle_bowl = "$7.75"

lunch = [tea, eggrolls, noodle_bowl]
lunch
```

```python
# Programs are composed of 5 big ideas/parts
# input, output, math, conditional execution, repetition
total = 0
for item in lunch:
  if type(item) == int or type(item) == float:
    total += item
  else:
    # Reassigning the item variable here
    item = item.replace("$", "")
    item = float(item)
    total += item
```
